## Задание 1

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
data = pd.read_csv("C:/Users/skaup/Python learning/Netologia/Python/notebooks/12. допфайлы/ml-latest-small/ratings.csv")
data
```

```python
def class_film(param):
    if param <=2:
        return ("низкий рейтинг")
    elif 2 < param <= 4:
        return ("средний рейтинг")
    else:
        return ("высокий рейтинг")

```

```python
data['class'] = data['rating'].apply(class_film)
data
```

```python
data['class'].value_counts()
```

geo_data = {

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

2.1. для каждой строки пройдите по всем годам списка years

2.2. если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска

2.3. если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
movies_df = pd.read_csv("C:/Users/skaup/Python learning/Netologia/Python/notebooks/12. допфайлы/ml-latest-small/movies.csv")
movies_df
```

```python
ratings_df = pd.read_csv("C:/Users/skaup/Python learning/Netologia/Python/notebooks/12. допфайлы/ml-latest-small/ratings.csv")
ratings_df
```

```python
# мёрджим:

x = ratings_df.merge(movies_df, how='left')
x
```

```python
# создаем таблицу индекс, название фильма, средний ретинг

films_rating = x.groupby('title').mean().reset_index()[['title', 'rating']]
films_rating
```

```python
# 1. В переменную years запишите список из всех годов с 1950 по 2010:

years = [str(i+1950) for i in range(61)]
years
```

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

2.1. для каждой строки пройдите по всем годам списка years

2.2. если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска

2.3. если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

```python
# создаем функцию с вариантами оценки: 

def production_year(row):
    for year in years:
        if year in row['title']:
            return year
    return '1900'
```

```python
# 3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец 'year':

films_rating['year'] = films_rating.apply(production_year, axis=1)
films_rating
```

```python
# 4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца 'year' и отсортируйте результат по убыванию рейтинга:

films_rating.sort_values('rating', ascending=False)
```
